```python
products_image = []
products_name = []
pricing = []
old_pricing = []
discount = []
rating = []

for content in prod_contents:
    for product in content:
        # Let's scrape the product image
        image = product.findAll("div", {"class": "img-c"})
        if len(image) == 0:
            products_image.append(np.nan)
        else:
            products_image.append(image[0].findAll("img",{"class":"img"})[0]['src'])
        
        # Let's scrape the product name
        name = product.findAll("h3", {"class": "name"})
        if len(name) == 0:
            products_name.append(np.nan)
        else:
            products_name.append(name[0].text)

        # scrape for price
        price = product.findAll("div", {"class": "prc"})
        if len(price) == 0:
            pricing.append(np.nan)
        else:
            pricing.append(price[0].text)

        # scrape for old price
        old_price = product.findAll("div", {"class": "old"})
        if len(old_price) == 0:
            old_pricing.append(np.nan)
        else:
            old_pricing.append(price[0].text)

        # Now let's have a look at its discount:
        discount_off = product.findAll("div", {"class": "bdg _dsct _sm"})
        if len(discount_off) == 0:
            discount.append(np.nan)
        else:
            discount.append(discount_off[0].text)

        
        # Now let's have a look at its free delivery:
        rating_star = product.findAll("div", {"class": "rev"})
        if len(rating_star) == 0:
            rating.append(np.nan)
        else:
            rating.append(rating_star[0].text)
           
print(len(products_image))
print(len(products_name))
print(len(pricing))
print(len(old_pricing))
print(len(discount))
print(len(rating))
```
